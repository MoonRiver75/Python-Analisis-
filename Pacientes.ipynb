```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

archivo_csv = 'CIJ_Pacientes1erIngreso_2019_2023T3.csv'
dataframe = pd.read_csv(archivo_csv)

# Consumo de todas las drogas
columnas_seleccionadas = ['ALCOHOL', 'MARIGUANA', 'COCAINA', 'INHALABLES', 'ESTIMULANT', 'ALUCINOGEN', 'DEPRESORES', 'OPIACEOS', 'OTROS']
datos_seleccionados = dataframe[columnas_seleccionadas]

# Filtrar por el consumo de al menos una droga
consumen_alguna_droga = datos_seleccionados[datos_seleccionados.sum(axis=1) > 0]

# Verificar la existencia de la columna 'SEXO' antes de continuar
if 'SEXO' in consumen_alguna_droga.columns:
    # Contar el consumo por género
    conteo_por_genero = consumen_alguna_droga.groupby('SEXO').size().reset_index(name='Total')

    # Crear el mapa de árbol por género para todas las drogas
    plt.figure(figsize=(10, 5))
    squarify.plot(sizes=conteo_por_genero['Total'], label=conteo_por_genero['SEXO'])
    plt.axis("off")
    plt.title('Consumo de Todas las Drogas por Género')
    plt.show()
else:
    print("La columna 'SEXO' no está presente en los datos filtrados.")

```

```python
columnas_drogas = ['ALCOHOL', 'MARIGUANA', 'COCAINA', 'INHALABLES', 'ESTIMULANT', 'ALUCINOGEN', 'DEPRESORES', 'OPIACEOS', 'OTROS']
for droga in columnas_drogas:
    consumen_droga = dataframe[dataframe[droga] == 1]
    conteo_por_genero_droga = consumen_droga.groupby('SEXO').size().reset_index(name='Total')
    plt.figure(figsize=(10, 5))
    squarify.plot(sizes=conteo_por_genero_droga['Total'], label=conteo_por_genero_droga['SEXO'])
    plt.axis("off")
    plt.title(f'Consumo de {droga} por Género')
    plt.show()
```

```python
dataframe['CONSUMO_TOTAL'] = dataframe[columnas_drogas].sum(axis=1)
consumen_total = dataframe[dataframe['CONSUMO_TOTAL'] > 0]
conteo_por_genero_total = consumen_total.groupby('SEXO').size().reset_index(name='Total')
plt.figure(figsize=(10, 5))
squarify.plot(sizes=conteo_por_genero_total['Total'], label=conteo_por_genero_total['SEXO'])
plt.axis("off")
plt.title('Consumo Total de Drogas por Género')
plt.show()
```

```python
columnas_drogas = ['ALCOHOL', 'MARIGUANA', 'COCAINA', 'INHALABLES', 'ESTIMULANT', 'ALUCINOGEN', 'DEPRESORES', 'OPIACEOS', 'OTROS']

dataframe['CONSUMO_TOTAL'] = dataframe[columnas_drogas].sum(axis=1)

consumen_total = dataframe[dataframe['CONSUMO_TOTAL'] > 0]

plt.figure(figsize=(12, 8))

for i, droga in enumerate(columnas_drogas):
    plt.subplot(3, 3, i+1)
    conteo_por_estado_conyugal_droga = consumen_total.groupby(['ESTADO_CONYUGAL', droga]).size().unstack().reset_index()
    plt.pie(conteo_por_estado_conyugal_droga[1], labels=conteo_por_estado_conyugal_droga['ESTADO_CONYUGAL'], autopct='%1.1f%%', colors=sns.color_palette("Oranges", n_colors=len(conteo_por_estado_conyugal_droga)))
    plt.title(f'Consumo de {droga} por Estado Conyugal')

plt.tight_layout()
plt.show()
```

```python
dataframe['CONSUMO_TOTAL'] = dataframe[columnas_drogas].sum(axis=1)
consumen_total = dataframe[dataframe['CONSUMO_TOTAL'] > 0]

plt.figure(figsize=(12, 8))
sns.violinplot(x='SEXO', y='CONSUMO_TOTAL', data=consumen_total, palette="Set2")
plt.xlabel('Género')
plt.ylabel('Consumo Total de Drogas')
plt.title('Distribución del Consumo Total de Drogas por Género')
plt.show()
```

```python
matriz_correlacion_drogas = consumen_total[columnas_drogas].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion_drogas, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Matriz de Correlación entre Sustancias')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
consumen_ambas = consumen_total[(consumen_total['COCAINA'] == 1) & (consumen_total['MARIGUANA'] == 1)]
model = LinearRegression()
X = consumen_ambas[['COCAINA']]
y = consumen_ambas['MARIGUANA']
model.fit(X, y)
X_pred = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
y_pred = model.predict(X_pred)
plt.figure(figsize=(12, 8))
sns.scatterplot(x='COCAINA', y='MARIGUANA', hue='ESTADO_CONYUGAL', data=consumen_ambas, palette='viridis')
plt.plot(X_pred, y_pred, color='red', linewidth=2, label='Línea de Tendencia')
plt.xlabel('Consumo de COCAINA')
plt.ylabel('Consumo de MARIGUANA')
plt.title('Relación entre Consumo de COCAINA y MARIGUANA por Estado Conyugal')
plt.legend()
plt.show()
```
